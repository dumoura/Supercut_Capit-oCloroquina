### Sampling video 

```python
for i in range(len(df)): 
    vid  = df.iloc[i, 0]
    start = df.iloc[i, 1]
    end = df.iloc[i, 2]
    msec = datetime.datetime.now().microsecond
    
    clip = VideoFileClip(f"{INPUT_DIR}/{vid[:-3]}.mp4", audio=True, target_resolution=None, resize_algorithm='bicubic').subclip(t_start = start, t_end=end)
    
    txt_vidID = TextClip(f"\n VideoId: {vid} \n", font="Futura bold", color='white', fontsize=18).set_pos(("left" , "top"))
    txt_timecode = TextClip(f"\n {start} - {end } \n", font="Futura Futura bold", color='white', fontsize=18).set_pos(("right" , "top"))
    
    txt_vidID =   txt_vidID .set_duration(2.5)
    txt_timecode =  txt_timecode.set_duration(2.5)
    
    final = CompositeVideoClip([clip, txt_vidID, txt_timecode])
    final.write_videofile(f"{CACHE_DIR}/{msec}_{vid}.mp4", preset="superfast", codec="libx264", audio_codec="aac", remove_temp=True, fps=25) #ultrafast, superfast, veryfast, faster, fast, medium,slow, slower, veryslow, placebo
    
    
```
