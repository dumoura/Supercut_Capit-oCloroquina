```python

def get_channel_videos(channel_id):
    
    # get Uploads playlist id
    
    res = youtube.channels().list(id=channel_id, 
                                  part='contentDetails').execute()
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    
    videos = []
    next_page_token = None
    
    while 1:
        res = youtube.playlistItems().list(playlistId=playlist_id, 
                                           part='snippet', 
                                           maxResults=50,
                                           pageToken=next_page_token).execute()
       
        videos += res['items']
        next_page_token = res.get('nextPageToken')
        
        if next_page_token is None:
            break
    
    return videos
```

```python
#channel_id = UCBrntWupyTRnnsRy3dA4VdA

videos = get_channel_videos('UCBrntWupyTRnnsRy3dA4VdA')
```

```python
video_ids = list(map(lambda x:x ['snippet']['resourceId']['videoId'], videos))

def get_vid_stats (video_ids):
    
    stats = []
    
    for i in range(0, len(video_ids), 50):
        res = youtube.videos().list(id=",".join(video_ids[i:i+50]),
                                   part='statistics').execute()
        
        stats += res['items']
    
    return stats
```

```python
stats = get_vid_stats(video_ids)
```

```python
for i in stats:
    print(i['statistics']['likeCount'], 
          i['statistics']['dislikeCount'],
          i['statistics']['commentCount'],
          i['id'],
          "https://www.youtube.com/watch?v="+i['id'])
```

```python
stats_cb = []

for i in stats:
    view = i['statistics']['viewCount']
    favorite = i['statistics']['favoriteCount']
    like = i['statistics']['likeCount'] 
    dislike = i['statistics']['dislikeCount']
    commentCount = i['statistics']['commentCount']
    Id = i['id']
    link = "https://www.youtube.com/watch?v="+i['id']
    stats_cb.append((view, favorite, like, dislike, commentCount, Id, link))
    

stats_cb = pd.DataFrame(stats_cb, columns = ['View','Favorite', 'Like', 'Dislike', 'Comment', "Id", 'Link'])
 
```

```python
comment = []

for r in comment_cb['items']:
    PublishedAt = r['snippet']['topLevelComment']['snippet']['publishedAt']
    Name = r['snippet']['topLevelComment']['snippet']['authorDisplayName']
    AuthorChannelUrl = r['snippet']['topLevelComment']['snippet']['authorChannelUrl']
    Like = r['snippet']['topLevelComment']['snippet']['likeCount']
    Reply = r['snippet']['totalReplyCount']
    Text = r['snippet']['topLevelComment']['snippet']['textOriginal']
    
    comment.append((PublishedAt, Name, AuthorChannelUrl, Like, Reply, Text))
```
