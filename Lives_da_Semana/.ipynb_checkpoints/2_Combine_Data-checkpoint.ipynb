```python
import pandas as pd
import os
```

```python
cwd = os.getcwd()
```

```python
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, "data")
CACHE_DIR = os.path.join(BASE_DIR, "cache")
os.makedirs(CACHE_DIR, exist_ok=True)
```

```python
for filename in os.listdir(DATA_DIR):
    print(filename)
```

```python
df = pd.read_csv(os.path.join(DATA_DIR, "2021.csv"))
```

```python
#somar [ou quaquer mat ] um valor a coluna 

df["Start_Year"] = df["Year"] + 10

df.head(2)
```

```python

#garantir o tipo de dados
csv_file = [x for x in os.listdir(DATA_DIR) if x.endswith(".csv")]

# total 

my_dataframes = []

#loop
for filename in csv_file:
    csv_path = os.path.join(DATA_DIR, filename)
    this_df = pd.read_csv(csv_path)
    #add a collumn with filename
    year = filename.replace(".csv", "")
    this_df['filename'] = filename
    this_df['year'] = year
    my_dataframes.append(this_df)
    print(this_df.head(n=1))
```

```python
my_dataframes
    
```

```python
# combine into one df

# DF vazia
# my_total_df = pd.DataFrame() => isso não funciona



my_total_df = pd.concat(my_dataframes)
my_total_df.head()
```

```python
dataset = os.path.join(CACHE_DIR, "movies-box-office-dataset.csv")
my_total_df.to_csv(dataset, index=False)
```
